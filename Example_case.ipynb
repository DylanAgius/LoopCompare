# An example of how to use loop compare

## Loading and fitting data

```python
import loopcompare as lc

#Firtsly load the information in the excel file which in this case is name 'hysteresis'
load=lc.load_info('hysteresis')

#Fit to these curves
#You can specify the number of x points you would like to use in the comparision via 'xpoints='
#and the number of polynomials to use in the fitting via 'poly='
load.fitcurves(xpoints=100,poly=4)
```

## Plotting the fit

You can visualise how good the fit is by plotting the results against the supplied data.

```python
#supply the hysteresis loop you interested in via 'interesloop='
#this is particular useful if you are comparing more than one loop
load.plotter(interestloop=1)
```

## Extracting the error

The percentage error will be supplied for each loop supplied in the excel file

```python
print(load.errorcal())
```
